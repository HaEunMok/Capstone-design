## 2. 당뇨병 데이터셋을 이용한 의사결정나무 분류 모델 만들기




### 데이터

- pima 인디언들의 당뇨병에 대한 정보. 서구화된 식습관으로 인한 당뇨 데이터
- 당뇨병 여부를 예츨할 수 있는 데이터
- 숫자로만 이루어져 전처리가 필요하지 않음


Pregnancies : 임신 횟수

Glucose : 2시간 동안의 경구 포도당 내성 검사에서 혈장 포도당 농도

BloodPressure : 이완기 혈압 (mm Hg)

SkinThickness : 삼두근 피부 주름 두께 (mm), 체지방을 추정하는데 사용되는 값

Insulin : 2시간 혈청 인슐린 (mu U / ml)

BMI : 체질량 지수 (체중kg / 키(m)^2)

DiabetesPedigreeFunction : 당뇨병 혈통 기능

Age : 나이

Outcome : 768개 중에 268개의 결과 클래스 변수(0 또는 1)는 1이고 나머지는 0입니다.

https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

### 라이브러리 로드

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

### 데이터셋 로드

```python
df= pd.read_csv("/content/drive/MyDrive/ML DL/diabetes.csv")
df.shape
```

```python
df.head()
```

### 학습, 예측 데이터 나누기

```python
# 8:2 비율로 split. 학습데이터 : 테스트데이터 = 8:2
split_count = int(df.shape[0] * 0.8)
split_count
```

```python
train = df[:split_count].copy()
train.shape
```

```python
test = df[split_count:].copy()
test.shape
```

### 학습, 예측에 사용할 칼럼 추출

```python
feature_names = train.columns[:-1].tolist()
feature_names
```

### 예측 칼럼 추출

```python
label_name = train.columns[-1]
label_name
```

## 학습, 예측 데이터셋 생성하기
- 학습데이터와 예측 데이터 행의 갯수가 일치하여야 한다

```python
X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
y_train = train[label_name]
print(y_train.shape)
y_train.head()
```

```python
X_test= test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
y_test = test[label_name]
print(y_test.shape)
y_test.head()
```

### 머신러닝 알고리즘 로드

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model
```

### 학습

```python
model.fit(X_train, y_train)
```

### 예측

```python
y_predict = model.predict(X_test)
y_predict[:5]
```

# 시각화

## 트리 알고리즘

```python
from sklearn.tree import plot_tree

plot_tree(model, feature_names=feature_names)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 20))
tree = plot_tree(model, feature_names=feature_names,
                           filled=True,
                           fontsize=10)
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_tree = export_graphviz(model, feature_names = feature_names,
filled=True)
graphviz.Source(dot_tree)
```

```python
# 피처의 중요도 추충
model.feature_importances_
```

```python
# 피처의 중요도 시각화
# 클루코스가 당뇨병 여부를 결정하는 핵심 피처이다.
sns.barplot(x=model.feature_importances_, y=feature_names)
```

## 정확도 예측하기

```python
diff_count = abs(y_test - y_predict).sum()
diff_count
```

```python
abs(y_test - y_predict).sum() / len(y_test) # 28% 잘못 예측했다
```

```python
(len(y_test) - diff_count) / len(y_test) * 100 # 71% 맞았다
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```

```python
model.score(X_test, y_test) * 100
```
